Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

Simple Linear Regression (SLR) is a basic statistical method used to understand the relationship between two variables:

One independent variable (X) ‚Üí the cause / input

One dependent variable (Y) ‚Üí the effect / output

In simple words, SLR helps us see how one thing changes when another thing changes.

To understand relationships
Example:

How does study time affect marks?

How does experience affect salary?

How does number of rooms affect the price of house?

To predict future values
If you know X, Simple Linear Regression helps you to estimate Y.

To measure impact
It tells us how strongly X affects Y (positive, negative, or no effect).

To simplify data
Instead of many data points, we summarize the pattern using one line.

Question 2: What are the key assumptions of Simple Linear Regression?

1. Linearity

There must be a linear relationship between the independent variable (X) and the dependent variable (Y).

2. Independence of Errors

The residuals (errors) should be independent of each other.
An error associated with one observation should not influence the error of another observation.

3. Homoscedasticity

The variance of the residuals should remain constant for all values of X.
In other words, the spread of errors should be uniform and not increase or decrease with X.

4. Normality of Errors

The residuals should be normally distributed with a mean of zero.

5. No Significant Outliers

There should be no extreme outliers that strongly affect the regression line.
Outliers can distort the slope and intercept of the model.

Question 5: What is the method of least squares in linear regression?

Method of Least Squares in Linear Regression :

The method of least squares is used to find the best-fit line in linear regression.

It works by minimizing the sum of the squares of errors between actual values and predicted values.

An error (residual) is the difference between the observed value and the value predicted by the line.

Squaring the errors ensures all values are positive and gives more weight to larger errors.

The line obtained using this method gives the most accurate predictions for the given data.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Logistic regression is a classification algorithm used to predict categorical outcomes.

The output is usually binary (e.g., Yes/No, Pass/Fail, 0/1).

It uses the sigmoid (logistic) function to convert values into probabilities between 0 and 1.

Example: Predicting whether an email is spam or not spam.

Linear regression is used to predict continuous numerical values, while logistic regression is used to predict categorical outcomes.

Linear regression produces outputs that can be any real number, but logistic regression outputs values only between 0 and 1.

Linear regression fits a straight line, whereas logistic regression fits an S-shaped curve.

Linear regression is used for problems like predicting salary, while logistic regression is used for problems like predicting spam or not spam.

Question 7: Name and briefly describe three common evaluation metrics for regression models.


Mean Absolute Error (MAE):

MAE measures the average absolute difference between the actual values and the values predicted by the regression model. It tells us, on average, how much the predictions are off from the true values, without considering the direction of the error. Since it does not square the errors, it treats all errors equally and is easy to understand.

Mean Squared Error (MSE):

MSE calculates the average of the squared differences between the actual and predicted values. By squaring the errors, it gives more importance to larger mistakes, making it useful when large errors are especially undesirable. However, the values can become large and harder to interpret because of squaring.

Root Mean Squared Error (RMSE):

RMSE is the square root of MSE and represents the average prediction error in the same units as the output variable. This makes it more interpretable than MSE while still penalizing larger errors. It is commonly used to compare different regression models.

Question 8: What is the purpose of the R-squared metric in regression analysis?

Purpose of the R-squared (R2) metric in regression analysis:

R-squared is used to measure how well a regression model explains the variability in the dependent variable.

It represents the proportion of total variation in the output variable that is explained by the input variable(s).

The value of R2 lies between 0 and 1.

An R2 value of 0 means the model explains none of the variation, while an R2
value of 1 means the model explains all the variation.

A higher R2 indicates a better fit of the regression model to the data.

It helps in comparing different regression models to see which one explains the data better.

```python
'''
Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)
'''# Simple Linear Regression using scikit-learn

import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X = years of experience
# y = salary
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([30000, 35000, 40000, 45000, 50000])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

In a simple linear regression model, the equation is:
y=mx+c

Slope (m or coefficient):
The slope represents the change in the dependent variable for a one-unit increase in the independent variable.
For example, if the slope is 5000 in a salary vs experience model, it means that for every additional year of experience, the salary increases by ‚Çπ5000.

Intercept (c):
The intercept is the value of the dependent variable when the independent variable is zero.
It indicates the starting point of the line. For example, if the intercept is ‚Çπ25,000, it represents the estimated salary when experience is zero.
